<h2> Data Visualisation: </h2>

To visualize distribution of data **regression line** is drawn over the points.

```python
sns.jointplot(x=df['x'], y=df['y'], data=df, kind='reg')
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(
    df['x'],
    df['y'],
    c='black'
)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
X = df['x'].values.reshape(-1,1)
y = df['y'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(
    df['x'],
    df['y'],
    c='black'
)
plt.plot(
    df['x'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

<h3> Why do we need to find the R² value? </h3>
<p>As for the R² metric, it measures the proportion of variability in the target that can be explained using a feature X. Therefore, assuming a linear relationship, if feature X can explain (predict) the target, then the proportion is high and the R² value will be close to 1. If the opposite is true, the R² value is then closer to 0.</p>

```python
X = df['x']
y = df['y']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

A R^2 value of 0.989 indicates that about 98% of the variability of 'x' is explained by 'y'.
